```python
import nltk

from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
#Word to vec
from gensim.models import Word2Vec
```

```python
import re
```

```python
paragraph="""Leadership was one of Gandhi's heroic qualities. One way he demonstrated leadership was by encouraging people in his village to start using homespun clothing. Homespun was one of Gandhi’s favorite hobbies. Introducing homespun to his village promised the growth of the village’s economy. The poor became employed to make homespun clothes for others in the village and outside of their own village. This drastically diminished the poverty in Gandhi’s village. Moreover, Gandhi ended the issue of discrimination against untouchables. Back in those days, the Hindus used to discriminate against the untouchables, who were their own kind but just low in the caste system. The untouchables are just another word to describe poor, less educated people who really didn’t fit in with the middle class people. Gandhi made the Hindus realize that what British people were doing to them is the same thing that they were doing to the untouchables. This made the Hindus realize what they were doing to the untouchables. Last but not least, the Indian Independence Movement was the all-time heroic deed led by Gandhi. This movement was accomplished in 1947 by Mahatma. This movement took place because India was governed by the British, and Gandhi wanted India to become independent. Therefore, Gandhi showcased many outstanding leadership skills that made him a hero."""
```

```python
#Preprocessing the data

text=re.sub(r'\[[0-9]*\]'," ",paragraph)
```

```python
text
```

```python
text=re.sub(r'\s+'," ",text)
```

```python
text
```

```python
text=text.lower()
text=re.sub(r'\d'," ",text)
text
```

```python
text=re.sub(r'\s+'," ",text)
text
```

```python
#Prepare the data
sentence=nltk.sent_tokenize(text)
sentence
```

```python
corpus=[]

for i in range(len(sentence)):
    words=nltk.word_tokenize(sentence[i])
    words=[word for word in words if word not in set(stopwords.words('english'))]
    corpus.append(words)
                                             
```

```python
corpus
```

```python
#Training the Word2Vec model

model=Word2Vec(corpus,min_count=1)
```

```python
model
```

```python
words=model.wv.vocab
```

```python
#Finding word vectors
vector=model.wv['way']
vector
```

```python
#Most similar words

similar=model.wv.most_similar('people')
similar
```
